```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns
import numpy as np
import pandas as pd

plt.figure(figsize=(5,1))
```

```python
df_n3245fy = pd.read_csv('n3245fy.csv')
df_n3245fy.set_index('chapter', inplace=True)
df_n3245fy_line = pd.read_csv('n3245fy_line.csv')
df_n3245fy_line.set_index(['chapter', 'line'], inplace=True)
df_n3245fy_all_line = pd.read_csv('n3245fy_all_line.csv')
df_n3245fy_all_line.set_index(['chapter', 'line'], inplace=True)
```

```python
df_n3245fy.plot.line(figsize=(10, 5), title='n3245fy', subplots=True, sharey=True)
plt.savefig('data/n3245fy.png')
```

```python
plt.close()
df_n3245fy_line_joban = df_n3245fy_line.loc[1:10]
df_n3245fy_line_joban.plot.line(figsize=(10, 5), title='n3245fy_dialog_only', subplots=True, sharey=True)
plt.savefig('data/n3245fy_dialog_only.png')
```

```python
plt.close()
df_n3245fy_all_line_joban = df_n3245fy_all_line.loc[1:2]
df_n3245fy_all_line_joban.plot.line(figsize=(10, 5), title='n3245fy_all_line', subplots=True, sharey=True)
plt.savefig('data/n3245fy_all_line.png')
```
